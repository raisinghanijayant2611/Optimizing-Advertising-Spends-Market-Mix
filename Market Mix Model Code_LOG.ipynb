# 1. Overall model

## 1.1.1 Overall Ordinary Linear Regression: TV, Online and Price_avg_model are significant

`TV` are postive and significant while `Online` and `price_avg_model` are negative and significant.

```python
## Year 1995 as dummy year and brand Audi as the brand 
import statsmodels.api as sm
X = sm.add_constant(X) # add constant to the regression
est = sm.OLS(Y, X)
est2 = est.fit()
print(est2.summary())
```

## 1.1.2 Overall Log-Log model: TV and Price_avg_model are significant

Interpretation: 
$\Delta y = \beta_{1} \Delta x$

After the log transformation to both X and Y, `TV` and `Price_avg_model` remain robust and `Online` loses its significance. e.g. Holding other factors constant, if `TV` channel increases its spending by 1%, the sales would increase 0.21%.

```python
# log-log model
import numpy as np
logX = X.copy()
logX[['TV','Print','Radio','Online','Outdoor','Prices_avg_model']]=logX[['TV','Print','Radio','Online','Outdoor','Prices_avg_model']].apply(lambda x: np.log(x+1))
logY = Y.apply(lambda y: np.log(y+1))

estlog = sm.OLS(logY, logX)
estlog2 = estlog.fit()
print(estlog2.summary())
```

## 1.2.1 OLS Channel Elasticity

```python
def elascoeff(val):
    X=channel_sales_df[[val]]
    Y=channel_sales_df[['Sales']]
    #Y.isna().sum()
    est = sm.OLS(Y, X)
    est2 = est.fit()
    d = est2.params.values[0]
    return(d)

## Channel Elasticity 
model1params = est2.params
data1 = channel_sales_df
data1 = data1[["TV",'Print', "Radio","Online","Outdoor", "Prices_avg_model"]]
data1mean = pd.DataFrame(data1.mean())
data1mean.columns = ["MeanValues"]
data1mean["Sales"] = channel_sales_df["Sales"].mean()
data1mean["Variance"] = data1.var()
data1mean

data1mean["channels"] = data1mean.index

data1mean["Coefficients"] = model1params
# data1mean["Coefficients"] = data1mean["channels"].map(elascoeff)
data1mean["Elasticity"] = data1mean["Coefficients"]*(data1mean["MeanValues"]/data1mean["Sales"])
data1mean
```

## 1.2.2 log-log model elasticity: not improve elasticity

$ log(Y) = \beta_{1}log(X)+ \beta_{0} $

$\beta_{1}$ IS THE ELASTICITY.

`Log-Log model doesn't make improvement to the elasticity.`

```python
coef_log = estlog2.params[estlog2.params.index.isin(['TV','Print','Radio','Online','Outdoor', "Prices_avg_model"])]
mean_logX = logX[['TV','Print','Radio','Online','Outdoor', "Prices_avg_model"]].mean()
var_logX = logX[['TV','Print','Radio','Online','Outdoor', "Prices_avg_model"]].var()
mean_logY  = logY[['Sales']].mean()


data1log = pd.DataFrame(data=[mean_logX, coef_log, var_logX])
data1log = data1log.T.rename(columns={0:'MeanValues', 1:'Coefficients', 2:'Variance'})
data1log['Sales'] = mean_logY[0]
data1log['channel'] = data1log.index
data1log = data1log[['MeanValues', 'Sales', 'Variance', 'channel', 'Coefficients']]
data1log["Elasticity"] = data1log["Coefficients"]
data1log
```
